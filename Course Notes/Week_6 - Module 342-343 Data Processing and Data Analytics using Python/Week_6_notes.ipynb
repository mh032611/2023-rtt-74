```python
file = open("spider.txt")
print(file.readline())
```

Keyword 'with' for opening and automatically closing.

```python
# When we use "with", all the sub commands are processed as if the file is opened and closed. 

# Using the readline method will produce only the first line because the file gets closed and then reopened with the 2nd with keyword  "with"
with open('spider.txt') as file:
    print(file.readline())
    
with open('spider.txt') as file: # each time you use the 'with' keyword it automatically opens and closes the file. So this will print the first and second line. 
    print(file.readline())
    print(file.readline())

```

Iterating through Files

```python
with open('spider.txt') as file:
    for line in file:
            print(line.upper())

print(type(line)) # <class 'str'>
print(type(file)) # <class '_io.TextIOWrapper'> 
                # this is a variable that is assigned for the open and closed function  
```
