## One-Dimensional Arrays

```python
# One-Dimensional Array - Example

# Creating a 1-dimensional NumPy array with 5 elements
my_array = np.array([1, 2, 3, 4, 5])
# Accessing elements of the array using indexing
print(my_array[0])    # Output: 1
print(my_array[3])    # Output: 4
```

```python
# Modifying elements of the array using indexing
my_array[2] = 7
print(my_array)       # Output: [1 2 7 4 5]
# Accessing a subset of the array using slicing
print(my_array[1:4])  # Output: [2 7 4]
# Modifying a subset of the array using slicing
my_array[1:4] = np.array([6, 8, 9])
print(my_array)
# Output: [1 6 8 9 5]
```

## Index and Slicing

```python
# Create a 1D numpy array
arr = np.array([1, 2, 3, 4, 5])
# Indexing: Access a specific element in the array
element = arr[2] # Returns the element at index 2 (which is 3)
print(element)
# Slicing: Access a range of elements in the array
slice = arr[1:4] # Returns elements at indices 1, 2, and 3
print(slice)
```

## Arithmetic operations

In this example, we created a one-dimensional numpy array arr with three elements. We then 
performed element-wise arithmetic operations with scalar values (1 and 2) using numpy.

```python
# Create a 1D numpy array
arr = np.array([1, 2, 3])
# Perform arithmetic operations
sum_arr = arr + 1 # Element-wise sum with scalar 1
prod_arr = arr * 2 # Element-wise product with scalar 2
div_arr = arr / 2 # Element-wise division with scalar 2
# Print the results
print(sum_arr)
print(prod_arr)
print(div_arr)
```

## Aggregate functions

In this example, we use Numpy's aggregate functions to calculate the sum of all elements in arr 
(arr_sum), the mean of all elements in arr (arr_mean), the minimum value in arr (arr_min), and 
the maximum value in arr (arr_max).

```python
# Create a 1D numpy array
arr = np.array([1, 2, 3, 4, 5])
# Perform aggregate functions
arr_sum = np.sum(arr) # Sum of all elements
arr_mean = np.mean(arr) # Mean of all elements
arr_min = np.min(arr) # Minimum value
arr_max = np.max(arr) # Maximum value
# Print the results
print(arr_sum, arr_mean, arr_min, arr_max, sep="," )
```

## Broadcasting

In this example, we create a 2D numpy array arr with dimensions 2x3. We then use 
numpy's broadcasting feature to multiply the entire array by a scalar value of 2. This 
operation multiplies each element in arr by 2, effectively doubling the values in the 
array.

```python
# Create a 2D numpy array
arr = np.array([[1, 2, 3], [4, 5, 6]])
# Multiply the array by a scalar value
result = arr * 2
# Print the result
print(result)
```

## Reshaping and Transposing

In this example, we create a 2D numpy array matrix with 2 rows and 3 columns. We then 
use numpy's T attribute to transpose the matrix, effectively swapping its rows and columns. 
We store the transposed matrix in a variable called transposed_matrix. We then use 
numpy's reshape function to reshape the transposed matrix into a 2D matrix with 3 rows 
and 2 columns. We store the reshaped matrix in a variable called reshaped_matrix.

```python
# Create a 2D numpy array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Transposing the array using the T attribute 
transposed_arr = arr.T


print(arr)
print(transposed_arr)
```

### Boolean Operations

Here are some examples of Boolean operations using Python's NumPy library:

a. Element-Wise comparison
b. Logical AND
c. Logical OR
d. Logical NOT
e. Array-wise comparison

#### Element-Wise Comparison

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 2, 3, 3, 5])
c = a == b

print(c) # Output: [False  True  True 
# False  True]
```

#### Logical AND


```python
a = np.array([True, True, False, False])
b = np.array([True, False, True, False])
c = np.logical_and(a, b)

print(c) # Output: [ True False False 
# False]

# The comparison checks to see if they are both True if both elements are true, the logical and will evaluate to true
# If one or more is false, the logical_and will evaluate to False
# Note True and False 
```

#### Logical OR

```python
a = np.array([True, True, False, False])
b = np.array([True, False, True, False])
c = np.logical_or(a, b)

print(c) # Output: [ True  True  True 
# False]

#This boolean comparison tests for a TRUE bool. If one or more of the array elements is true, the logical_or evaluates to True
```

#### Logical NOT

```python
a = np.array([True, False, True, False])
b = np.logical_not(a)

print(b) # Output: [False  True False  
# True]
```

#### Array-Wise Comparison

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 2, 3, 3, 5])
c = np.array_equal(a, b)

# Compares corresponding elements with each array (they have to be same length, same size, etc)

print(c) # Output: False
```

#### Sorting

In this example, we first create a NumPy array (a) with some random values. We then use 
the np.sort() function to sort the array in ascending order and store the sorted array in a 
new variable b. Finally, we print the sorted array (b).

```python
a = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3])
b = np.sort(a)

print(b) # Output: [1 1 2 3 3 4 5 5 6 9]
```

#### Masking   

In this example, we first create a NumPy array (a) with some random values. We then create 
a boolean mask by comparing each element in the array (a) with the value 3. The result is a 
boolean array with True values, where the condition is satisfied and False values, where it is 
not. We then use this mask to select only the elements of (a) where the mask is True, and 
store them in a new variable (b). Finally, we print the masked array (b).

```python
a = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3])
mask = a > 3
b = a[mask]

print(b) # Output: [4 5 9 6 5]


print(type(b)) # <class 'numpy.ndarray'>
print(type(a)) # <class 'numpy.ndarray'>
print(type(mask)) # <class 'numpy.ndarray'>


# The word "mask" has nothing to do with the function. you can do it without using mask 
a = np.array([1,2,3,4,5,6])
b = a[a>3]

print(b)
```
